```
import re
import time
import requests
import numpy as np
from textblob import TextBlob, Word
from scrapy.http import TextResponse
import pandas as pd
```

```
url = "https://www.imdb.com/chart/top"
page = requests.get(url)
response = TextResponse (url=page.url, body=page.text, encoding="utf-8")
```

```
name =response.css("td[class='titleColumn'] a::attr(title)").extract()
```

```
name
```

```
year = response.css("span[class='secondaryInfo']::text").re("\d{4}")
```

```
year
```

```
rating = response.css("td[class = 'ratingColumn imdbRating'] >strong::text").extract()
```

```
rating
```

```
before_1990 = [i for i in year if int(i) < 1990]
```

```
len(before_1990)
```

```
movie_dict = {"name":name, "year":year, "rating":rating}
```

```
movie_df = pd.DataFrame(movie_dict)
```
